```python
#@title กด 1 ครั้งก่อนเริ่มโปรแกรม { display-mode: "form" }
import pickle
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
loaded_model = pickle.load(open('/content/booking_model.sav', 'rb'))
```

```python
#@title Run เพื่อกรอกข้อมูลโรงแรมของคุณ { display-mode: "form" }
facility_score=float(input('คะแนนสิ่งอำนวยความสะดวก :'))
comfort_score=float(input('คะแนนความสะดวกสบาย :'))
value_score=float(input('คะแนนความคุ้มค่า :'))
score=[facility_score,comfort_score,value_score]
```

```python
#@title Result { display-mode: "form" }
price=loaded_model.predict([score])[0]
if price > 100:
  print(f"เราคาดว่าราคาของห้องพักนี้ประมาณ : {price}")
else:
  print(f"ขออภัย model ขอเราไม่สามารถทำนายข้อมูลนี้ได้")
```
